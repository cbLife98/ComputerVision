```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images//uni.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap = 'gray')
```

```python
# Create a filter for edge detection(3x3)
# Sobel Filter
sobel_x = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])

# Perform convolution using Filter2D, which has inputs: (grayscale,kernel,
# bit-depth(output type[-1 means same as input]

filtered_image = cv2.filter2D(gray,-1,sobel_x)

plt.imshow(filtered_image,cmap = 'gray')
```

```python
# Binary Image where the strong edges captured above are clearly visible above a certain threshold
retval,binary_image = cv2.threshold(filtered_image,100,255,cv2.THRESH_BINARY)
plt.imshow(binary_image,cmap = 'gray')
```
