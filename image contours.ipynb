```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images//hand.jpg')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
#  binary inverse makes image white and background black
retval,binary = cv2.threshold(gray,225,255,cv2.THRESH_BINARY_INV)
plt.imshow(binary,cmap = 'gray')
```

```python
# heirarchy is for finding contours in contours , returns list of contours
contours,hierarchy = cv2.findContours(binary,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

image_c = np.copy(image_copy)
#image,contour list, all[-1],color,width of boundary
all_contours = cv2.drawContours(image_c,contours,-1,(0,255,0),2)
plt.imshow(all_contours)
```
