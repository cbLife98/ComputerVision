```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images/waffle.jpg')

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)

plt.imshow(gray,cmap = 'gray')
```

```python
# Harris Accept floats values
gray = np.float32(gray)

# image,window size,kernel,how much to acceot as corner
dst = cv2.cornerHarris(gray,2,3,0,0.04)

dst = cv2.dilate(dst,None)


plt.imshow(dst,cmap = 'gray')
```

```python
thresh = 0.08 * dst.max()

corner_image = np.copy(image)

for j in range(0,dst.shape[0]):
    for i in range(0,dst.shape[1]):
        if(dst[j,i]>thresh):
            # image ,center point, radius,color,thickness
            cv2.circle(corner_image,(i,j),2,(0,255,0),1)
plt.imshow(corner_image)
```
