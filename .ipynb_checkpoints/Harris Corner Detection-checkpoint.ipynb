```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline
```

```python
image = cv2.imread('images//chess.png')

image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
#harris takes a float array
gray = np.float32(gray)
#destination = image,size of window(2,2),kernel size = 3,3,which points to consider corner(if high more corners)
dst = cv2.cornerHarris(gray,2,3,0.04)
# corners as bright points and non as darker enhance the points 
dst = cv2.dilate(dst,None)
plt.imshow(dst,cmap = 'gray')
```

```python
thresh = 0.01 * dst.max()#if less then more corners identified
corner_image = np.copy(image_copy)

for j in range(0,dst.shape[0]): #x-cordinate
    for i in range(0,dst.shape[1]):#y-cordinate
        if(dst[j,i]>thresh):
            # image,center pt,radius,color,thickness
            cv2.circle(corner_image,(i,j),2,(0,255,0),1)
plt.imshow(corner_image)
```
