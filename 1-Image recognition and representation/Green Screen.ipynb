```python
# Imports needed 
import numpy as np
import matplotlib.pyplot as plt
import cv2   # Computer vision library

%matplotlib inline
```

```python
# Read the image
image = cv2.imread('images/car_green_screen.jpg')

# Dimensions of the image 
print("The Dimensions of the image are: ",image.shape)
```

```python
# Make a copy of the image
image_copy = np.copy(image)

# Change BGR to RGB
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

# Diplay the image
plt.imshow(image_copy)
```

```python
# Define color range which will be blocked else will be black (play around with this)
lower = np.array([0,150,0])
upper = np.array([80,255,80])
```

```python
# Define masked area
mask = cv2.inRange(image_copy,lower,upper)

# Display
plt.imshow(mask,cmap = 'gray')
```

```python
# To show this image and instead block the background 
mask_copy = np.copy(image_copy)

# Show the pizza
mask_copy[mask != 0]  = [0,0,0]

# Display
plt.imshow(mask_copy)
```

```python
# Add a new image 
back = cv2.imread('images/space_background.jpg')
back = cv2.cvtColor(back,cv2.COLOR_BGR2RGB)

# Crop it to the size of image
crop_back = back[ 0:450, 0:660 ]

# Mask the cropped background so that pizza is blocked
crop_back[mask == 0] = [0,0,0]

plt.imshow(crop_back)
```
