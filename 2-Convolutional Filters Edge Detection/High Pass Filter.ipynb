```python
# Imports needed 
import numpy as np
import matplotlib.pyplot as plt
import cv2   # Computer vision library

%matplotlib inline
```

```python
# Read the image
image = cv2.imread('images/city_hall.jpg')

# Dimensions of the image 
print("The Dimensions of the image are: ",image.shape)

# Make a copy of the image
image_copy = np.copy(image)

# Change BGR to RGB
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

# Diplay the image
plt.imshow(image_copy)
```

```python
# Change  RGB to gray
gray = cv2.cvtColor(image_copy,cv2.COLOR_BGR2GRAY)

# Diplay the image
plt.imshow(gray,cmap='gray')
```

```python
# 3x3 kernel also called sobel filter
sobel_x = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])
# Applying this filter to the image (image,bit-depth,kernel)
filtered_image = cv2.filter2D(gray,-1,sobel_x)

# Display it 
plt.imshow(filtered_image,cmap = 'gray')
```

```python
# Convert it to binary to see the lines above a certain threshold
retval,binary_image = cv2.threshold(filtered_image,50,255,cv2.THRESH_BINARY)

# Display it
plt.imshow(binary_image,cmap = 'gray')
```
