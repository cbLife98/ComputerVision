```python
import numpy as np
import matplotlib.pyplot as plt
import cv2


%matplotlib inline
```

```python
image = cv2.imread('images\\brain.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap = 'gray')
```

```python
# Applying Gaussian Blur
guass = cv2.GaussianBlur(gray,(5,5),0)
plt.imshow(guass,cmap = 'gray')
```

```python
sobel_x = np.array([[-1,0,1],
                   [-2,0,2],
                   [-1,0,1]])
```

```python
both_apply = cv2.filter2D(guass,-1,sobel_x)
one_apply = cv2.filter2D(gray,-1,sobel_x)

f, (ax1,ax2) = plt.subplots(1,2,figsize =(20,20))

ax1.set_title('Low Pass')
ax1.imshow(both_apply,cmap = 'gray')

ax2.set_title('High Pass')
ax2.imshow(one_apply,cmap = 'gray')
```

```python
retval1,binary_image1 = cv2.threshold(both_apply,50,255,cv2.THRESH_BINARY)
retval,binary_image2 = cv2.threshold(one_apply,50,255,cv2.THRESH_BINARY)

f, (ax1,ax2) = plt.subplots(1,2,figsize =(20,20))

ax1.set_title('Low Pass')
ax1.imshow(binary_image1,cmap = 'gray')

ax2.set_title('High Pass')
ax2.imshow(binary_image2,cmap = 'gray')
```
