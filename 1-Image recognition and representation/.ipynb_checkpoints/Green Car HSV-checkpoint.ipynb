```python
# Imports needed 
import numpy as np
import matplotlib.pyplot as plt
import cv2   # Computer vision library

%matplotlib inline
```

```python
# Read the image
image = cv2.imread('images/car_green_screen2.jpg')

# Dimensions of the image 
print("The Dimensions of the image are: ",image.shape)
```

```python
# Make a copy of the image
image_copy = np.copy(image)

# Change BGR to RGB
image_copy = cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)

# Diplay the image
plt.imshow(image_copy)
```

```python
# TO extract a single color we need to plot the channels seperately r g b and h s v
r = image_copy[:,:,0]
g = image_copy[:,:,1]
b = image_copy[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize =(20,10))

ax1.set_title("RED")
ax1.imshow(r,cmap = 'gray')

ax2.set_title("GREEN")
ax2.imshow(g,cmap = 'gray')

ax3.set_title("BLUE")
ax3.imshow(b,cmap = 'gray')
```

```python
# Converting RGB to HSV 
hsv = cv2.cvtColor(image_copy,cv2.COLOR_RGB2HSV)

# Setting parameter for hsv
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1,ax2,ax3) = plt.subplots(1,3,figsize =(20,10))

ax1.set_title("HUE")
ax1.imshow(h,cmap = 'gray')

ax2.set_title("SATURATION")
ax2.imshow(s,cmap = 'gray')

ax3.set_title("VALUE")
ax3.imshow(v,cmap = 'gray')
```

```python
# Define our color selection boundaries in HSV values

## TODO: Change these thresholds
# This initial threshold allows a certain low range for Hue (H)
lower_hue = np.array([0,0,0]) 
upper_hue = np.array([100,255,255])

# Define the masked area
mask = cv2.inRange(image_copy, lower_hue, upper_hue)

# Mask the image to let the car show through
masked_image = np.copy(image_copy)

masked_image[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```
